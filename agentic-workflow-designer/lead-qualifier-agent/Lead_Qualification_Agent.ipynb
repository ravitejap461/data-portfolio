```python
import pandas as pd

# Simulated inbound lead data
data = [
    {
        "Name": "Ravi Teja",
        "Role": "CTO",
        "Company": "Finverse.ai",
        "Message": "Looking to integrate your fraud detection API into our platform.",
        "Website": "finverse.ai",
        "LinkedIn": "https://linkedin.com/in/raviteja"
    },
    {
        "Name": "Anjali Mehra",
        "Role": "Marketing Executive",
        "Company": "GrowFast",
        "Message": "Just exploring what's out there. Can I get a demo?",
        "Website": "growfast.com",
        "LinkedIn": "https://linkedin.com/in/anjalimehra"
    },
    {
        "Name": "Raj Patel",
        "Role": "Head of Procurement",
        "Company": "InfraCore",
        "Message": "We want to automate supplier onboarding using your tool.",
        "Website": "infracore.in",
        "LinkedIn": "https://linkedin.com/in/rajpatel"
    },
]

df = pd.DataFrame(data)
df
```

```python
import random

def simulate_llm_qualification(row):
    message = row['Message'].lower()
    role = row['Role'].lower()

    if 'integrate' in message or 'automate' in message or 'API' in message or 'platform' in message:
        tag = "Hot"
        reason = "Strong product interest and technical role"
        reply = f"Hi {row['Name']}, thanks for reaching out! Based on your role and message, we’d love to schedule a deep dive to explore how we can integrate with {row['Company']}."
    elif 'demo' in message or 'exploring' in message:
        tag = "Warm"
        reason = "General curiosity, non-decision maker role"
        reply = f"Hi {row['Name']}, happy to show you around. Here's a link to book a demo — feel free to invite relevant team members from {row['Company']}."
    else:
        tag = "Cold"
        reason = "Unclear intent"
        reply = f"Hi {row['Name']}, thanks for your interest in our offerings. Can you help us understand what you're looking to achieve at {row['Company']}?"

    return pd.Series([tag, reason, reply])

# Apply simulated LLM classification and response
df[['LeadScoreTag', 'LLMReasoning', 'SuggestedResponse']] = df.apply(simulate_llm_qualification, axis=1)

df.head()
```
