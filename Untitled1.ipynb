```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the uploaded file
df = pd.read_csv("churn_data.csv")

# Show the first few rows
df.head()
```

```python
print("Churn Distribution:")
print(df['Churned'].value_counts())
```

```python
churn_rate = df['Churned'].mean() * 100
print(f"\nChurn Rate: {churn_rate:.1f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Churned', y='TenureMonths')
plt.title("Tenure vs. Churn")
plt.xlabel("Churned (0 = Stayed, 1 = Left)")
plt.ylabel("Months as a Customer")
plt.show()
```

```python
sns.boxplot(data=df,x='Churned',y='SupportTicketsRaised')
plt.title("SupportTicketsRaised vs. Churn")
plt.xlabel("Churned (0 = Stayed, 1 = Left)")
plt.ylabel("SupportTicketsRaised")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Churned', y='MonthlyUsageHours')
plt.title("MonthlyUsageHours vs. Churn")
plt.xlabel("Churned")
plt.ylabel("MonthlyUsageHours")
plt.show()
```
