```python
import pandas as pd

df = pd.read_csv("Sample - Superstore.csv", encoding="ISO-8859-1")
print(df.shape)
df.head()
```

```python
df.groupby("Category")[["Sales", "Profit"]].sum().sort_values(by="Profit", ascending=False)
```

```python
df.groupby("Region")[["Sales", "Profit"]].sum().sort_values(by="Profit", ascending=False)
```

```python
df.groupby("Sub-Category")[["Profit"]].sum().sort_values(by="Profit", ascending=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="Discount", y="Profit", data=df)
plt.title("Discount vs Profit")
plt.show()
```
